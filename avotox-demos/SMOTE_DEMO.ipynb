```python
import pandas as pd
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
```

```python
# Imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=21, n_classes=2, weights=[0.91, 0.08], random_state=42)
```

```python
# Visualize the class distribution before SMOTE
plt.figure(figsize=(8, 6))
plt.hist(y, bins=2, color='m', ec='k')
plt.title("Class Distribution Before SMOTE")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

Use 'SMOTE' from the 'imblearn' library to generate synthetic samples for the minority class

```python
# Apply Smote
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
# Visualize the class distribution after SMOTE
plt.figure(figsize=(8, 6))
plt.hist(y_res, bins=2, color='m', ec='k')
plt.title("Class Distribution After SMOTE")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```
